#2.matplotlib 라이브러리 자유자재로 사용하기

2-1 기초 그래프 그리기

```python
tips=sns.load_dataset("tips")
print(tips.head())
print(type(tips))
```

```python
fig=plt.figure()
axes1=fig.add_subplot(1,1,1)
```

```python
axes1.hist(tips['total_bill'],bins=10) #x축 간격을 10으로
axes1.set_title('Histogram of Total Bill')#제목 붙이기
axes1.set_xlabel('Frequency') #X축 이름
axes1.set_ylabel('Total Bill') #Y축 이름

fig
```



*   산점도 그래프: 이변량 그래프
  * 격자.scatter(x데이터,y데이터)



```python
scatter_plot=plt.figure()
axes1=scatter_plot.add_subplot(1,1,1)
axes1.scatter(tips['total_bill'], tips['tip']) #axes1.plot(tips['total_bill'], tips['tip'],'o')동일
axes1.set_title('Scatterplot of Total Bill vs Tip') 
axes1.set_xlabel('Total Bill') 
axes1.set_ylabel('Tip')
```



*   박스 그래프
  * 격자.boxplot( )


```python
boxplot = plt.figure() 
axes1 = boxplot.add_subplot(1, 1, 1) 

axes1.boxplot( 
    [tips[tips['sex'] == 'Female']['tip'], 
     tips[tips['sex'] == 'Male']['tip']], 
    labels=['Female', 'Male'])


axes1.set_xlabel('Sex') 
axes1.set_ylabel('Tip') 
axes1.set_title('Boxplot of Tips by Sex')1
```

2-2 다변량 그래프 그리기



*   새 변수 추가



```python
def recode_sex(sex):
    if sex == 'Female':
        return 0 
    else:
        return 1
```

```python
tips['sex_color']=tips['sex'].apply(recode_sex)
```



*   색깔은 c 사이즈는 s



```python
scatter_plot = plt.figure() 
axes1 = scatter_plot.add_subplot(1, 1, 1) 
axes1.scatter(
    x=tips['total_bill'], 
    y=tips['tip'],    
    s=tips['size'] * 10,    
    c=tips['sex_color'],
    alpha=0.5)
axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size') 
axes1.set_xlabel('Total Bill') 
axes1.set_ylabel('Tip') 
```

#5.seaborn 라이브러리로 그래프 스타일 설정하기

set_style 메서드를 사용하여 바꿀 수 있고 한번 실행하면 그 이후에도 모두 지정한 스타일로 바뀜

```python
fig, ax = plt.subplots() 
ax = sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True) 
```



*   whitegrid:바깥 배경에 줄 긋기



```python
sns.set_style('whitegrid') 
fig, ax = plt.subplots() 
ax = sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True) 
```



*   for문 사용하여 한번에 다 적용해보기



```python
fig = plt.figure() 
seaborn_styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks'] 
for idx, style in enumerate(seaborn_styles):
    plot_position = idx + 1
    with sns.axes_style(style):
        ax = fig.add_subplot(2, 3, plot_position)
        violin = sns.violinplot(x='time', y='total_bill', data=tips, ax=ax)
        violin.set_title(style) 
        
fig.tight_layout() 
```
